```python
#Age distribution
age_groups = ['Under 30', '30-39', '40-49', '50-59', '60-69', '70+']
age_bins = [0, 30, 40, 50, 60, 70, c14['age'].max() + 1]
age_group_labels = [f'{age_groups[i]} ({age_bins[i]}-{age_bins[i+1]-1})' for i in range(len(age_groups))]
c14['age_group'] = pd.cut(c14['age'], bins=age_bins, labels=age_group_labels)
c15['age_group'] = pd.cut(c15['age'], bins=age_bins, labels=age_group_labels)
age_group_counts_2014 = c14['age_group'].value_counts().reindex(age_group_labels, fill_value=0)
age_group_counts_2015 = c15['age_group'].value_counts().reindex(age_group_labels, fill_value=0)
plt.figure(figsize=(12, 6))
width = 0.4
x = range(len(age_group_labels))
plt.bar([i - width/2 for i in x], age_group_counts_2014, width=width, label='2014', color='skyblue')
plt.bar([i + width/2 for i in x], age_group_counts_2015, width=width, label='2015', color='lightcoral')
plt.title('Age Distribution of Candidates (2014 vs. 2015)')
plt.xlabel('Age Group')
plt.ylabel('Number of Winners')
plt.xticks(x, age_group_labels, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group the data by state for both 2014 and 2015
state_data_2014 = c14.groupby('state')
state_data_2015 = c15.groupby('state')

# Calculate the total number of candidates for each state for both years
total_candidates_2014 = state_data_2014['name'].count()
total_candidates_2015 = state_data_2015['name'].count()

# Calculate the total number of winners for each state for both years
total_winners_2014 = state_data_2014['result'].sum()
total_winners_2015 = state_data_2015['result'].sum()

# Calculate the percentage of winners relative to the number of candidates for both years
percentage_winners_2014 = (total_winners_2014 / total_candidates_2014) * 100
percentage_winners_2015 = (total_winners_2015 / total_candidates_2015) * 100

# Create a grouped bar chart to visualize the distribution
plt.figure(figsize=(12, 6))
width = 0.4
x = range(len(percentage_winners_2014))
plt.bar([i - width/2 for i in x], percentage_winners_2014, width=width, label='2014', color='skyblue')
plt.bar([i + width/2 for i in x], percentage_winners_2015, width=width, label='2015', color='lightcoral')
plt.title('Percentage of Winners Relative to Candidates by State (2014 vs. 2015)')
plt.xlabel('State')
plt.ylabel('Percentage of Winners (%)')
plt.xticks(x, percentage_winners_2014.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Voter Tunrout Rate by State
state_data_2014 = r14.groupby('state').agg({
    'undi_keluar_peti': 'sum',
    'pengundi_jumlah': 'sum'
})
state_data_2014['percentage_2014'] = (state_data_2014['undi_keluar_peti'] / state_data_2014['pengundi_jumlah']) * 100
state_data_2015 = r15.groupby('state').agg({
    'undi_keluar_peti': 'sum',
    'pengundi_jumlah': 'sum'
})
state_data_2015['percentage_2015'] = (state_data_2015['undi_keluar_peti'] / state_data_2015['pengundi_jumlah']) * 100
plt.figure(figsize=(12, 6))
width = 0.4
x = range(len(state_data_2014))
plt.bar([i - width/2 for i in x], state_data_2014['percentage_2014'], width=width, label='2014', color='skyblue')
plt.bar([i + width/2 for i in x], state_data_2015['percentage_2015'], width=width, label='2015', color='lightcoral')
plt.title('Voter Turnout Rate by State (2014 vs. 2015)')
plt.xlabel('State')
plt.ylabel('Percentage (%)')
plt.xticks(x, state_data_2014.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Percentage of Invalid Votes by State 
state_data_2014 = r14.groupby('state').agg({
    'undi_rosak': 'sum',
    'undi_keluar_peti': 'sum'
})
state_data_2014['percentage_2014'] = (state_data_2014['undi_rosak'] / state_data_2014['undi_keluar_peti']) * 100
state_data_2015 = r15.groupby('state').agg({
    'undi_rosak': 'sum',
    'undi_keluar_peti': 'sum'
})
state_data_2015['percentage_2015'] = (state_data_2015['undi_rosak'] / state_data_2015['undi_keluar_peti']) * 100
plt.figure(figsize=(12, 6))
width = 0.4
x = range(len(state_data_2014))
plt.bar([i - width/2 for i in x], state_data_2014['percentage_2014'], width=width, label='2014', color='skyblue')
plt.bar([i + width/2 for i in x], state_data_2015['percentage_2015'], width=width, label='2015', color='lightcoral')
plt.title('Percentage of Invalid Votes by State (2014 vs. 2015)')
plt.xlabel('State')
plt.ylabel('Percentage (%)')
plt.xticks(x, state_data_2014.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import chi2_contingency

categorical_columns = ['state', 'ethnicity', 'party']
result_column = 'result'
chi_square_results = []
for column in categorical_columns:
    contingency_table_2014 = pd.crosstab(c14[column], c14[result_column])
    chi2_2014, p_2014, _, _ = chi2_contingency(contingency_table_2014)
    contingency_table_2015 = pd.crosstab(c15[column], c15[result_column])
    chi2_2015, p_2015, _, _ = chi2_contingency(contingency_table_2015)
    chi_square_results.append([column, chi2_2014, chi2_2015,p_2014, p_2015])

chi_square_results_df = pd.DataFrame(chi_square_results, columns=['Column', 'Chi-squared(2014)', 'Chi-squared(2015)', 'p-value(2014)', 'p-value(2015)'])
print(chi_square_results_df.to_string(index=False))
```
